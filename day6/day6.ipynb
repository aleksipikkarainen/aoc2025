<h3> Day 6 - Part 1 - Problem</h3>
<p>We have a file with five large lists, four full of numbers and one full of either asterisks or pluses. We have to perform a calculation by taking the four values from each list at each index and using the same index, performing the operation described in the final list. Then we combine the totals into our answer.</p>

<h3> Day 6 - Part 1 - Answer</h3>
<p> We read the five lists into respective arrays depending on the contents that we catch with regex. Then we transpose the arrays into individual calculations, extract the components required for the calculation and just iterate through our unique calculations while performing them.</p>

```python
import re
file = open("day6input25.txt")
calculations = []

for list_ in file:
    
    if "+" in list_:
        operators = re.findall(r'\+|\*',list_)
        calculations.append(operators)
        
    else:
        numbers = re.findall(r'\d+',list_)
        calculations.append(numbers)

transposed_calculations = [[calculations[j][i] for j in range(len(calculations))] for i in range(len(calculations[0]))]
total = 0

for calculation in transposed_calculations:
    sign = calculation[-1:][0]
    start = int(calculation[0]) 
    numbers = calculation[1:4]

    for number in numbers:
        
        if sign == '+':
            start += int(number)
        else:
            start *= int(number)
            
    total += start
    
print(total)
```
