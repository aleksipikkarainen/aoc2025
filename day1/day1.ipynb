<h3> Day 1 - Part 1 - Problem</h3>

<p>
We have a file with directions such as "R23" and "L45" that move a dial right and left x steps respectively. The dial loops on itself. How many times does it hit 0?</p>

<h3> Day 1 - Part 1 - Answer </h3>

<p> So, we need to simulate a moving dial that loops on itself. Or do we? We only need to acquire information when the dial reaches zero. I see an easier way to find out:</p>

```python
print(100%100)
print(0%100)
print(-100%100)
```

<p> As you can see, by using the modulo we don't actually have to simulate a dial, instead we can just do a regular integer. Now all we have to do is parse the file and count the occurrences of zero.</p>

```python
f = open("day1input25.txt")
start = 50
password = 0

for x in f:
    turns = x[1:len(x)]

    if x[0] == "R":
        start += int(turns)

    else:
        start -= int(turns)

    if start%100 == 0:
        password += 1

print(password)
```

<p> And that was the correct answer! Onwards to part! </p>

<h3> Day 1 - Part 2 - Problem</h3>

<p> Additionally, how many times does the dial also pass 0?
</p>

<h3> Day 1 - Part 2 - Answer</h3>

<p> Well, our first part solution cannot directly take this into account. Now we need to do some comparisons between the moves to see how many times we move through zeroes and hundreds. </p>

<p> We can do this easily by for example: if we were to turn R256 when we are at 50: 
<ul>
    <li>Since 256 is over 100, we can just take the 2 in the hundreds place, add it to our password</li>
    <li>Then with the 56, we just run it like normally and additionally check if we cross a 100 mark.</li>
</ul>

There are four possible occurrences here: Existing number both positive and negative or new turn right or left, let's simulate what we would need to do:

<ul>
    <li>Existing 90, turn R20: We need to see if existing + turn > 100</li>
    <li>Existing -10, turn R20: We need to see if existing + turn > 0</li>
    <li>Existing 10, turn L20: We need to see if existing - turn < 0</li>
    <li>Existing -90 turn L20, We need to see if existing - turn < 100</li>
</ul>

These rules get a bit complicated, so let's move a step at a time and see when we hit a zero

</p>

```python
f = open("day1input25.txt")
start = 50
password = 0

for x in f:
    turns = x[1:len(x)]

    if x[0] == "R":
        for i in range(int(turns)):
            start += 1
            
            if start%100 == 0:
                password += 1

    else:
        for i in range(int(turns)):
            start -= 1
            
            if start%100 == 0:
                password += 1

print(password)
```

<p> And that's our correct answer, I would have liked to do this more efficiently but ran into some niche edge cases, sometimes simple is better at the cost of some computing power.</p>
