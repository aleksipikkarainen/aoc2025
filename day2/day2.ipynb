<p> In a list with specified ranges, how many times does a series of numbers repeat twice within one of the number values?
</p>

<h3>Day 2 - Part 1 - Answer</h3>

<p>Now this is a simple one, we just look at all of the numbers between the ranges, split them at the half and check if they match.</p>

<p> Let's just do a simple check to split the numbers at exactly half.</p>

```python
test = '123456' 
test2 = '11112222'

first = test[0:int(len(test)/2)]
print(first)
last = test[int(len(test)/2):len(test)] 
print(last)

first = test2[0:int(len(test2)/2)]
print(first)
last = test2[int(len(test2)/2):len(test2)] 
print(last)
```

<p> Now for the actual part:</p>

```python
password = 0

for x in ranges:
    range1, range2 = x.split("-")

    for i in range(int(range1), int(range2)+1):
        string = str(i) 
        first = string[0:int(len(string)/2)]
        second = string[int(len(string)/2):len(string)] 

        if first == second:
            password += int(string)
            
print(password)
```

<h3>Day 2 - Part 2 - Problem</h3>

<p>Now, how many times does a number within the range contain a repeating number pattern?
</p>

<h3>Day 2 - Part 2 - Answer</h3>
<p>Now we have to perform multiple splices depending on the string length. So we have to check whether a pattern repeats up to half of the strings length.
</p>

```python
password = 0

for x in ranges:
    range1, range2 = x.split("-")

    for i in range(int(range1), int(range2)+1):
        string = str(i)

        for k in range(1, int(len(string)//2)+1):
            if len(string)%k == 0:
                if string[0:k]*(len(string)//k) == string:
                    password += int(string) 
                    break
            
print(password)
```
