<p>This method should work, let's start:</p>

```python
#Let's open our file
f = open("day2input25.txt") 
#Turn our input into an array using the split()-method
for x in f:
    ranges = x.split(",")
```

```python
print(ranges)
```

<p> Now for the actual part:</p>

```python
#Initializing our password
password = 0
#Iterating through our ranges
for x in ranges:
    #Splitting the range from "-"
    range1, range2 = x.split("-")
    #Checking all values between the range, edges included
    for i in range(int(range1), int(range2)+1): #end range is not inclusive so +1
        #Converting our values to string in order to split
        string = str(i) 
        #Performing the splits
        first = string[0:int(len(string)/2)]
        second = string[int(len(string)/2):len(string)] 
        #Checking if the id is invalid
        if first == second:
            #Adding invalid id to the password
            password += int(string)
            
print(password)
```

<h3>Day 2 - Part 2 - Answer</h3>
<p>Now we have to perform multiple splices depending on the string length. So we have to check whether a pattern repeats up to half of the strings length.
</p>

```python
#Initializing our password
password = 0
#Iterating through our ranges
for x in ranges:
    #Splitting the range from "-"
    range1, range2 = x.split("-")
    #Checking all values between the range, edges included
    for i in range(int(range1), int(range2)+1): #end range is not inclusive so +1
        #Converting our values to string in order to split
        string = str(i)
        #Checking all possible substring lengths.
        for k in range(1, int(len(string)//2)+1):
            #Checking if the string can be divided by the range
            if len(string)%k == 0:
                #Checking if the string can be expressed as a multiplication of the substing
                if string[0:k]*(len(string)//k) == string:
                    #Adding our correct answer and breaking the loop
                    password += int(string) 
                    break
            
print(password)
```
