<h3> Day 3 - Part 1 - Question </h3>

<p> You descend a short staircase, enter the surprisingly vast lobby, and are quickly cleared by the security checkpoint. When you get to the main elevators, however, you discover that each one has a red light above it: they're all offline.

"Sorry about that," an Elf apologizes as she tinkers with a nearby control panel. "Some kind of electrical surge seems to have fried them. I'll try to get them online soon."

You explain your need to get further underground. "Well, you could at least take the escalator down to the printing department, not that you'd get much further than that without the elevators working. That is, you could if the escalator weren't also offline."

"But, don't worry! It's not fried; it just needs power. Maybe you can get it running while I keep working on the elevators."

There are batteries nearby that can supply emergency power to the escalator for just such an occasion. The batteries are each labeled with their joltage rating, a value from 1 to 9. You make a note of their joltage ratings (your puzzle input). For example:

987654321111111<br>
811111111111119<br>
234234234234278<br>
818181911112111<br>

The batteries are arranged into banks; each line of digits in your input corresponds to a single bank of batteries. Within each bank, you need to turn on exactly two batteries; the joltage that the bank produces is equal to the number formed by the digits on the batteries you've turned on. For example, if you have a bank like 12345 and you turn on batteries 2 and 4, the bank would produce 24 jolts. (You cannot rearrange batteries.)

You'll need to find the largest possible joltage each bank can produce. In the above example:
<ul>
    <li>In 987654321111111, you can make the largest joltage possible, 98, by turning on the first two batteries.</li>
    <li>In 811111111111119, you can make the largest joltage possible by turning on the batteries labeled 8 and 9, producing 89 jolts.</li>
    <li>In 234234234234278, you can make 78 by turning on the last two batteries (marked 7 and 8)</li>
    <li>In 818181911112111, the largest joltage you can produce is 92.</li>
</ul>
The total output joltage is the sum of the maximum joltage from each bank, so in this example, the total output joltage is 98 + 89 + 78 + 92 = 357.

There are many batteries in front of you. Find the maximum joltage possible from each bank; what is the total output joltage?
</p>

<h3> Day 3 - Part 1 - Answer </h3> 

<p> The first part seems to be fairly simple, we iterate through the lines recording two values in a way, that finds the largest possible value combinations from each line.</p>

```python
#Opening our file 
f = open("day3input25.txt")
#Taking the total sum
total = 0
#Iterating through our rows
for x in f:
    #Converting our "nextline" character into a dummy one for some legroom
    x = x.replace("\n","1")
    #Our starting value is the first one
    first = int(x[0]) 
    #A dummy 1 value which is the minimum value
    second = 1
    #Iterating through our numbers
    for i in range(1,len(x)):
        #If the current value is larger than the first and we are not at the end (eg. we cannot take the last number here)
        if int(x[i]) > first and i < len(x)-2:
            #Replacing the old values with the newer larger ones
            first = int(x[i])
            second = int(x[i+1])
        #If the current value is larger than second (here we can be at the end and also miss the dummy 1 at the end)
        elif int(x[i]) > second and i < len(x)-1: 
            #Replace the second value
            second = int(x[i])
    #Since our 'first' value represents the tens we multiply it before combining into our total with the second value
    total += first*10+second

print(total)
```
