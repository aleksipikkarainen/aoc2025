<h3> Day 3 - Part 1 - Problem </h3>

<p> In strings full of integers, what is the largest numerical value within when we assume that the tens value must come before the ones value?
</p>

<h3> Day 3 - Part 1 - Answer </h3> 

<p> The first part seems to be fairly simple, we iterate through the lines recording two values in a way, that finds the largest possible value combinations from each line.</p>

```python
f = open("day3input25.txt")
total = 0

for x in f:
    x = x.replace("\n","1")
    first = int(x[0]) 
    second = 1
    
    for i in range(1,len(x)):
        if int(x[i]) > first and i < len(x)-2:
            first = int(x[i])
            second = int(x[i+1])

        elif int(x[i]) > second and i < len(x)-1: 
            second = int(x[i])
            
    total += first*10+second

print(total)
```
