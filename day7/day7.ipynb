<h1> Advent of Code 2025 - Day 7</h1>
<h3>Day 7 - Part 1 - Problem</h3>
<p>I have a matrix with "^" characters sprinkled around as to form a triangle. At the top from "S" character, we must find out how many times a line that flows down and always splits into two at the "^" spots, splits in total.</p>

<h3> Day 7 - Part 1 - Answer</h3>
<p> I utilize a helper array and while iterating the file, each occurrence of "^" is recorded and the indices next to it are marked as "1", and the index we hit it with as "0". This way I can simulate where the line flows downwards. I then just record how many times we have hit the "^" signs in total.</p>

```python
file = open("day7input25.txt")
answer = 0
for row in file:
    
    if 'S' in row:
        start_pos = row.index('S')
        ways = [0] * len(row)
        ways[start_pos] = 1
        
    if '^' in row:
        for index in range(0,len(row)):
            if row[index] == '^':
                if ways[index] != 0:
                    ways[index+1] = 1
                    ways[index-1] = 1
                    ways[index] = 0
                    answer += 1
    
print(answer)
```

<h3> Day 7 - Part 2 - Problem</h3>
<p>Now, I must track how many different possibilities the line can take while flowing down in total.</p>

<h3> Day 7 - Part 2 - Answer</h3>
<p> Instead of marking where the line goes simply, I mark how many times the line has already been at a certain spot. Then when hitting another "^" sign, I subtract that current index to zero and add </p>

```python
file = open("day7input25.txt")
answer = 0
for row in file:
    
    if 'S' in row:
        start_pos = row.index('S')
        ways = [0] * len(row)
        ways[start_pos] = 1
        
    if '^' in row:
        for index in range(0,len(row)):
            if row[index] == '^':
                if ways[index] != 0:
                    total = ways[index]
                    ways[index+1] += total
                    ways[index-1] += total
                    ways[index] = 0



answer = sum(ways)    
print(answer)
```
